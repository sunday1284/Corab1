```python
# 판다스 모듈 탑재
import pandas as pd

# 승하차 인원 CSV 파일 읽어오기
data = pd.read_csv('/content/deajeon-logs1.csv',encoding='cp949')
# 데이터 확인하기
data
```

```python
# 지하철 역 정보 CSV 파일 읽어오기
station_info = pd.read_csv("/content/deajeon-logs2.csv", encoding='cp949')

# 데이터 확인하기
station_info
```

```python
# 승하차 인원 데이터 상세 정보 확인하기
data.head(10)
```

```python
# 색인 변경하기
station_info = station_info.set_index('역번호')

# 데이터 확인하기
station_info.set_index
```

```python
# 출근용 데이터 추출
morning_data = data[pd.to_datetime(data.timestamp).dt.hour < 09]
morning_data = morning_data[pd.to_datetime(data.timestamp).dt.hour > 06]
morning_data
```

```python
# 퇴근용 데이터 추출
evening_data = data[pd.to_datetime(data.timestamp).dt.hour > 17]
evening_data = evening_data[pd.to_datetime(data.timestamp).dt.hour < 21]
evening_data
```

```python
# 역 기준으로 그룹화하여 데이터 더하기
morning_station_sum = morning_data.groupby("station_code").sum()
evening_station_sum = evening_data.groupby("station_code").sum()
```

```python
# 역 정보 합치기
morning_joined_data = morning_station_sum.join(station_info)
evening_joined_data = evening_station_sum.join(station_info)
```

```python
# 출근 시간 승차 인원
morning_deajeon_in = folium.Map(location=[36.39340, 127.4030], zoom_start = 12)
HeatMap(data = morning_joined_data[['geo.', 'geo.longitude', 'people_in']]).add_to(morning_deajeon_in)
morning_deajeon_in
```

```python
# 출근 시간 하차 인원
morning_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(morning_seoul_out)
morning_seoul_out
```

```python
# 퇴근 시간 승차 인원
evening_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(evening_seoul_in)
evening_seoul_in
```

```python
# 퇴근 시간 하차 인원
evening_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(evening_seoul_out)
evening_seoul_out
```
